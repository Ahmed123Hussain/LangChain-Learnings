```
import pandas as pd
df = pd.read_excel("sample.xlsx")
df.head()
```

```
#dropping last 30 rows from the Dataframe
df2 = df.drop(df.tail(30).index,inplace = False)
df2
```

```
#Converting textual data in dataframe into vector embeddings.
#converting the manager column into vector(numbers) - 15 columns -> with shape (15,768)
!pip install pandas
from sentence_transformers import SentenceTransformer
encoder = SentenceTransformer("all-mpnet-base-v2")
vectors = encoder.encode(df2.Manager)
vectors.shape
```

```
vectors
```

```
#Creaing Index for FAISS DB
import faiss
index = faiss.IndexFlatL2(768)
index.add(vectors)
```

```
search_query = "Which country is an ally of US in middle east"
#now convert it into vector for similarity search
import numpy as np
vec = encoder.encode(search_query)

svec = np.array(vec).reshape(1, -1)
svec.shape
```

```
#searching for answer in the index (the vecctor with highest similarity is chosen)
index.search(svec, k=2) #k = no. of nearest vectors
```
